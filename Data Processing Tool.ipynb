### American Waste and Recycling Data

#### For Python analysis

````python
sheets_dict = pd.read_excel('data/MATERIALS_MUNICIPAL_WASTE_STREAM_1960_TO_2015.xlsx', sheet_name=None)

waste_df = pd.DataFrame()
for name, sheet in sheets_dict.items():
    sheet['Process Type'] = name
    waste_df = waste_df.append(sheet)
waste_df.head()
waste_df.reset_index(inplace=True, drop=True)
waste_df = waste_df.fillna('')
waste_df['Material'] = waste_df['Materials'] + waste_df['Products']
waste_df.drop(columns=['Materials', 'Products'], inplace=True)
waste_df = waste_df[waste_df.Material.str.contains('Total') == False]
waste_df = waste_df[waste_df.Material.str.contains('Subtotal') == False]
waste_df = waste_df[waste_df.Material.str.contains('Table') == False]
waste_df = waste_df[waste_df.Material != '']
waste_df['Material'] = waste_df.Material.str.replace(r'â€“', '-')
waste_df[['Material', 'Subcategory 1', 'Subcategory 2', 'Subcategory 3', 'Subcategory 4']] = waste_df['Material'].str.split('-', expand=True)
waste_df.reset_index(inplace=True, drop=True)
waste_df = waste_df.set_index(['Process Type', 'Material', 'Subcategory 1', 'Subcategory 2', 'Subcategory 3', 'Subcategory 4'])
waste_df.reset_index(inplace=True)
waste_df.head()
#waste_df.to_excel('waste_data.xlsx')
````
